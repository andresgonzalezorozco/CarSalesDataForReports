#**Client**

```python
class Client:

  def __init__(self, clientID, countryID, clientName, address1, address2, town,
               county, postCode, region, outerPostCode, clientType, clientSize,
               clientSince, isCreditWorthy, isDealer):
    self.clientID = clientID
    self.clientName = clientName
    self.address1 = address1
    self.address2 = address2
    self.town = town
    self.county = county
    self.postCode = postCode
    self.region = region
    self.outerPostCode = outerPostCode
    self.countryID = countryID
    self.clientType = clientType
    self.clientSize = clientSize
    self.clientSince = clientSince
    self.isCreditWorthy = isCreditWorthy
    self.isDealer = isDealer

  def getClientID(self):
    return clientID

  def setClientID(self, clientID):
    self.clientID = clientID

  def getClientName(self):
    return clientName

  def setClientName(self, clientName):
    self.clientName = clientName

  def getAddress1(self):
    return address1

  def setAddress1(self, address1):
    self.address1 = address1

  def getAddress1(self):
    return address2

  def setAddress1(self, address2):
    self.address2 = address2

  def getTown(self):
    return town

  def setTown(self, town):
    self.town = town

  def getCounty(self):
    return county

  def setCounty(self, county):
    self.county = county

  def getPostCode(self):
    return postCode
  
  def setPostCode(self, postCode):
    self.postCode = postCode
  
  def getRegion(self):
    return region

  def setRegion(self, region):
    self.region = region

  def getOuterPostCode(self):
    return outerPostCode

  def setOuterPostCode(self, outerPostCode):
    self.outerPostCode = outerPostCode

  def getCountryID(self):
    return countryID

  def setCountryID(self, countryID):
    self.countryID = countryID

  def getClientType(self):
    return clientType

  def setClientType(self, clientType):
    self.clientType = clientType

  def getClientSize(self):
    return clientSize
  
  def setClientSize(self, clientSize):
    self.clientSize = clientSize

  def getClientSinse(self):
    return clientSince

  def setClientSince(self, clientSince):
    self.clientSince = clientSince

  def getIsCreditWorthy(self):
    return isCreditWorthy

  def setIsCreditWorthy(self, isCreditWorthy):
    self.isCreditWorthy = isCreditWorthy

  def getIsDealer(self):
    return isDealer

  def setIsDealer(self,isDealer):
    self.isDealer = isDealer

  def toString(self): 
    print('Client:')
    print('clientID: ', self.clientID )
    print('countryId: ', self.countryID)
    print('clientName: ', self.clientName )
    print('address1: ', self.address1)
    print('address2: ', self.address2)
    print('town: ', self.town)
    print('county: ', self.county)
    print('postCode: ', self.postCode)
    print('region: ', self.region)
    print('outerPostCode: ', self.outerPostCode)
    print('clientType: ', self.clientType)
    print('clientSize: ', self.clientSize)
    print('clientSince: ', self.clientSince)
    print('isCreditWorthy: ', self.isCreditWorthy)
    print('isDealer: ', self.isDealer)
```

```python
class Stock:

  def __init__(self, stockID, make, model, colorID, vehicleType, costPrice,
               spareParts, laborCost, registrationDate, mileage, purchaseDate,
               vehicleAgeInYears):
    self.stockID = stockID
    self.make = make
    self.model = model
    self.colorID = colorID
    self.vehicleType = vehicleType
    self.costPrice = costPrice
    self.spareParts = spareParts
    self.laborCost = laborCost
    self.registrationDate = registrationDate
    self.mileage = mileage
    self.purchaseDate = purchaseDate
    self.vehicleAgeInYears = vehicleAgeInYears
  
  def getStockID(self):
    return stockID

  def setStockID (self, stockID):
    self.stockID = stockID

  def getMake(self):
    return make

  def setMake(self, make):
    self.make = make

  def getModel(self):
    return model

  def setModerl(self, model):
    self.model = model
  
  def getColorID(self):
    return colorID

  def setColorID(self, colorID):
    self.colorID = colorID
  
  def getVehicleType(self):
    return vehicleType
  
  def setVehicleType(self, vehicleType):
    self.vehicleType = vehicleType

  def getCostPrice(self):
    return costPrice

  def setCostPrice(self, costPrice):
    self.costPrice = costPrice
  
  def getSpareParts(self):
    return spareParts
  
  def setSpareParts(self, spareParts):
    self.spareParts = spareParts
  
  def getLaborCost(self):
    return laborCost

  def setLaborCost(self, laborCost):
    self.laborCost = laborCost
  
  def getRegistrationDate(self):
    return registrationDate
  
  def setRegistrationDate(self, registrationDate):
    self.registrationDate = registrationDate
  
  def getMileage(self):
    return mileage

  def setMileage(self, mileage):
    self.mileage = mileage

  def getPurchaseDate(self):
    return purchaseDate

  def setPurchaseDate(self, purchaseDate):
    self.purchaseDate = purchaseDate

  def getVehicleAgeInYears(self):
    return vehicleAgeInYears

  def setVehicleAgeInYears(self, vehicleAgeInYears):
    self.vehicleAgeInYears = vehicleAgeInYears

  def toString(self): 
    print('Stock: ')
    print('stockID: ', self.stockID)
    print('make: ', self.make)
    print('model: ', self.model)
    print('colorID: ', self.colorID)
    print('type: ', self.vehicleType)
    print('price: ', self.costPrice)
    print('spareParts: ', self.spareParts)
    print('laborCost: ', self.laborCost)
    print('registrationDate: ', self.registrationDate)
    print('mileage: ', self.mileage)
    print('purchaseDate: ', self.purchaseDate)
    print('vehicleAgeInYears: ', self.vehicleAgeInYears)
```
